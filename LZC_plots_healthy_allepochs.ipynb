```python
import mne
import numpy as np
import os

# List of file paths for the healthy subjects' .edf files
file_paths = [
    '/content/SC4102E0-PSG.edf',
    '/content/SC4292G0-PSG.edf',
    '/content/SC4401E0-PSG.edf',
    '/content/SC4512E0-PSG.edf',
    '/content/SC4561F0-PSG.edf',
    '/content/SC4631E0-PSG.edf',
    '/content/SC4702E0-PSG.edf',
    '/content/SC4761E0-PSG.edf',
]

def check_file_exists(file_path):
    """Check if the file exists."""
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"EDF file not found: {file_path}")
    print(f"File {file_path} found. Proceeding with loading.")

def load_data(file_path):
    """Load EEG data from an EDF file."""
    try:
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=True)
        print(f"File {file_path} loaded successfully.")
        return raw
    except Exception as e:
        print(f"Error loading file {file_path}: {e}")
        return None

def handle_annotations(raw):
    """Handle annotations and 'boundary' events in EDF files."""
    annotations = raw.annotations
    print(f"Annotations found in {raw.filenames[0]}: {annotations}")

    if annotations:
        boundary_indices = [i for i, desc in enumerate(annotations.description) if 'boundary' in desc]
        print(f"'Boundary' event indices: {boundary_indices}")

        if boundary_indices:
            valid_data = []
            start_idx = 0
            sfreq = raw.info['sfreq']

            for boundary in boundary_indices:
                boundary_start = int(annotations.onset[boundary] * sfreq)
                valid_data.append(raw[:, start_idx:boundary_start][0])
                start_idx = boundary_start

            valid_data.append(raw[:, start_idx:][0])
            corrected_data = np.concatenate(valid_data, axis=1)
            raw_corrected = mne.io.RawArray(corrected_data, raw.info)
            raw = raw_corrected
            print("Discontinuities handled and corrected.")
    return raw

def handle_nan_values(raw):
    """Handle NaN values and bad channels."""
    if np.any(np.isnan(raw.get_data())):
        print("NaN values found in the data. Interpolating bad channels...")
        raw.info['bads'] = ['EEG 001']  # Update with the actual bad channels if known
        raw.interpolate_bads()

def save_corrected_data(raw, corrected_file_path):
    """Save the corrected data."""
    raw.save(corrected_file_path, overwrite=True)
    print(f"Corrected file saved as: {corrected_file_path}")

def process_edf_files(file_paths):
    """Process all EDF files."""
    for file_path in file_paths:
        try:
            check_file_exists(file_path)
            raw = load_data(file_path)

            if raw is not None:
                # Handle annotations and boundary events
                raw = handle_annotations(raw)

                # Handle NaN values and bad channels
                handle_nan_values(raw)

                # Save the corrected data
                corrected_file_path = file_path.replace('.edf', '_corrected.fif')
                save_corrected_data(raw, corrected_file_path)

                # Optionally, plot the data to verify its integrity
                raw.plot()

        except Exception as e:
            print(f"An error occurred while processing {file_path}: {e}")

if __name__ == "__main__":
    process_edf_files(file_paths)
```

```python
import mne

# Define filter parameters
l_freq = 0.5  # Low cutoff frequency (Hz)
h_freq = 45.0  # High cutoff frequency (Hz)

# List of file paths for the 8 healthy subjects' .fif files
file_paths = [
    'subject_1_raw.fif',
    'subject_2_raw.fif',
    'subject_3_raw.fif',
    'subject_4_raw.fif',
    'subject_5_raw.fif',
    'subject_6_raw.fif',
    'subject_7_raw.fif',
    'subject_8_raw.fif',
]

# Process each subject
try:
    for i, file_path in enumerate(file_paths, start=1):
        # Load the .fif file
        raw = mne.io.read_raw_fif(file_path, preload=True)
        print(f"\nDataset {i} loaded successfully from {file_path}.")

        # Apply the bandpass filter
        raw_filtered = raw.copy().filter(
            l_freq=l_freq, h_freq=h_freq, method='iir',
            iir_params=dict(order=8, ftype='butter')
        )
        print(f"Subject {i}'s data filtered with 0.5-45 Hz bandpass.")

        # Optionally, save the filtered data
        filtered_file_path = file_path.replace('_raw.fif', '_filtered_raw.fif')
        raw_filtered.save(filtered_file_path, overwrite=True)
        print(f"Filtered dataset for Subject {i} saved as {filtered_file_path}")

except FileNotFoundError as fnfe:
    print(f"File not found: {fnfe}")
except Exception as e:
    print("An unexpected error occurred.")
    print(e)
```

```python
import numpy as np
from mne import make_fixed_length_epochs
import mne

# Define epoch length (30 seconds)
epoch_length = 30.0  # seconds

# List of file paths for the 8 healthy subjects' filtered .fif files
file_paths_filtered = [
    'subject_1_filtered_raw.fif',
    'subject_2_filtered_raw.fif',
    'subject_3_filtered_raw.fif',
    'subject_4_filtered_raw.fif',
    'subject_5_filtered_raw.fif',
    'subject_6_filtered_raw.fif',
    'subject_7_filtered_raw.fif',
    'subject_8_filtered_raw.fif',
]

# Process each subject
for i, file_path in enumerate(file_paths_filtered, start=1):
    try:
        # Load the filtered .fif file
        raw_filtered = mne.io.read_raw_fif(file_path, preload=True)
        print(f"\nDataset {i} loaded successfully from {file_path}.")

        # Segment the filtered data into 30-second epochs
        epochs = make_fixed_length_epochs(raw_filtered, duration=epoch_length, preload=True)
        print(f"Segmented Subject {i} into {len(epochs)} epochs.")

        # Save the epochs data with proper naming convention and overwrite if file exists
        epochs_file_path = file_path.replace('_filtered_raw.fif', '-epo.fif')  # Adjusted naming convention
        epochs.save(epochs_file_path, overwrite=True)
        print(f"Epochs for Subject {i} saved as {epochs_file_path}")

    except FileNotFoundError as fnfe:
        print(f"File not found: {fnfe}")
    except Exception as e:
        print(f"An unexpected error occurred for Subject {i}.")
        print(e)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne

def calculate_power_ratios(epochs):
    """Calculate power spectral ratios for each epoch."""
    # Define frequency bands
    bands = {
        'delta': (0.5, 4),
        'theta': (4, 8),
        'alpha': (8, 12),
        'beta': (12, 30),
        'gamma': (30, 45)
    }

    # Compute PSD for all epochs at once
    spectrum = epochs.compute_psd(method='welch', n_fft=512, n_overlap=256)  # Use smaller n_fft (512) and n_overlap (256)

    # Get the PSD data and frequency bins
    psds = spectrum.get_data()  # Shape: (n_epochs, n_channels, n_freqs)
    freqs = spectrum.freqs

    # Calculate power ratios for each band
    power_ratios = {}
    for band_name, (low_freq, high_freq) in bands.items():
        band_mask = (freqs >= low_freq) & (freqs <= high_freq)
        band_power = np.trapz(psds[:, :, band_mask], axis=2)  # Integrate across frequencies
        power_ratios[band_name] = band_power

    # Normalize power ratios to get relative power
    total_power = np.sum([np.sum(band_power, axis=1) for band_power in power_ratios.values()], axis=0)
    normalized_ratios = {band: band_power / total_power[:, None] for band, band_power in power_ratios.items()}

    return normalized_ratios

def binarize_signal(signal, threshold=None):
    """Binarize a signal based on its mean or a specified threshold."""
    if threshold is None:
        threshold = np.mean(signal)
    return (signal > threshold).astype(int)

def lempel_ziv_complexity(binary_signal):
    """Calculate the Lempel-Ziv complexity of a binary signal."""
    n = len(binary_signal)
    i, lz_complexity = 0, 1
    substrings = set()
    current_substring = ""

    while i < n:
        current_substring += str(binary_signal[i])
        if current_substring not in substrings:
            substrings.add(current_substring)
            current_substring = ""
            lz_complexity += 1
        i += 1

    return lz_complexity

def analyze_subject_data(epochs):
    """Analyze EEG data for power ratios and Lempel-Ziv complexity."""
    # Calculate power spectral ratios
    power_ratios = calculate_power_ratios(epochs)

    # Get all epoch data (shape: n_epochs x n_channels x n_times)
    all_epoch_data = epochs.get_data()

    # Calculate Lempel-Ziv complexity for each epoch
    lz_complexities = []
    for epoch_data in all_epoch_data:
        # Use the first channel (or modify as needed, e.g., average across channels)
        signal = epoch_data[0]
        binary_signal = binarize_signal(signal)
        lz_complexity_value = lempel_ziv_complexity(binary_signal)
        lz_complexities.append(lz_complexity_value)

    return power_ratios, lz_complexities

def plot_lzc_values(lz_complexities, subject_id, title="Lempel-Ziv Complexity (LZC)"):
    """Plot Lempel-Ziv Complexity (LZC) values."""
    plt.figure(figsize=(10, 6))
    plt.plot(lz_complexities, marker='o', linestyle='-', color='b')
    plt.title(f"{title} for Subject {subject_id}")
    plt.xlabel("Epoch Index")
    plt.ylabel("LZC Value")
    plt.grid()
    plt.show()

# Simulate data for 8 subjects
sfreq = 100  # Sampling frequency in Hz
n_channels = 7  # Adjust number of channels based on your data
rng = np.random.RandomState(42)

n_subjects = 8

# Specify the number of epochs for each subject
epochs_per_subject = {
    1: 2858,  # Subject 1
    2: 2810,  # Subject 2
    3: 2630,  # Subject 3
    4: 2750,  # Subject 4
    5: 2704,  # Subject 5
    6: 2758,  # Subject 6
    7: 2624,  # Subject 7
    8: 2630   # Subject 8
}

n_times = int(sfreq * 8)  # 8 seconds of data per epoch

# Create channel names based on the number of channels
ch_names = [f'EEG{i+1}' for i in range(n_channels)]
info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=['eeg'] * n_channels)

# Store results for all subjects
all_power_ratios = {}
all_lz_complexities = {}

# Analyze data for all 8 subjects
for subject_id in range(1, n_subjects + 1):
    n_epochs = epochs_per_subject[subject_id]  # Get the number of epochs for the current subject

    # Simulate data for each subject (replace with actual data loading if available)
    data_subject = rng.randn(n_epochs, n_channels, n_times)

    # Create EpochsArray object for each subject
    epochs = mne.EpochsArray(data_subject, info)

    # Analyze data for the current subject
    power_ratios, lz_complexities = analyze_subject_data(epochs)

    # Store results
    all_power_ratios[subject_id] = power_ratios
    all_lz_complexities[subject_id] = lz_complexities

    # Plot LZC values for the current subject
    plot_lzc_values(lz_complexities, subject_id, title="LZC Values for All Epochs")

# Optional: Display or save the power ratios and LZC values for each subject
# For example, print the power ratios for the first subject:
print("Power Ratios for Subject 1:", all_power_ratios[1])

```
